# Taller regresión lineal Univariada

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Vectores globales

```python
car=np.array([1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])# Voltaje
et=np.array([0.169610272,0.283395813,0.386358738,0.470227872,0.433281294,0.600267648,0.73833898,0.79031502,0.877464268,0.843564462,0.964438917])# Corriente
par=np.array([0.1,0.00021])#Me los invento inicialmente los primeros a probar
```

```python
"""
A nuestra funcín de costo le entran:
par: son los parámetros por ahora solo son dos o0 y o1 en ese orden,
car: nuestro vector numpy.ndarray, es X
et: nuestro vector de etiquetas es Y
"""
def costo(par,car,et):
  n1=len(car)
  n2=len(et)
  if(not (n1 == n2)):
    print("ERROR")
  n=n1
  o0=par[0]
  o1=par[1]
  J=(1/(2*n))*np.sum( np.power( ( (o0+o1*car)-(et) ) ,2) )
  return J
```

```python
print("El costo es: ",costo(par,car,et))
```

```python
#Aqui inicia nuestra busqueda exaustiva "fuerza bruta" "grid search"
#for 1 para iterar en o0

lista_par=[]
lista_costos=[]

for o0 in np.arange(0,0.1,0.001):
  #for 2 para iterar en o1
  for o1 in np.arange(0,0.2,0.001):
    #tomar o0 y o1 y meterlos en un array tupla
    parametros=(o0,o1)
    #prueban su hipotesis usando las funcion de costo que acaban de crear
    lista_costos.append(costo(parametros,car,et))
    #Guardan el valor del costo y los parámetros con los que se obtuvo ese costo
    lista_par.append(parametros)
    #print(lista_costos)
    #print(lista_par)
# miran cual conjunto de parámetros tuvo el mejor costo. 
# AYUDA miran la funcion ARGMIN del paquete np "numpy"
```

```python
plt.plot(lista_costos)
plt.show()
```

```python
print(np.min(lista_costos))
print(np.argmin(lista_costos))
par=lista_par[np.argmin(lista_costos)]
print(par)
#Yest=par[0]+par[1]*car+par[2]*car*car
Yest=0.046+0.157*car
plt.plot(car,Yest)
plt.plot(car,et)
plt.show()
```
