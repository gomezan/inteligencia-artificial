## Primer punto Distribuciones discretas.
1. Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números: 
1. Grafique los números aleatorios, 
2. Ordene de menor a mayor y grafique,
4. Encuentre los valores teoricos de media y varianza, y comparelos con los estimados.

```
#Para todas las simulaciones:
N=1000
m=300
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```
# bernoulli con cierto p entre 0 y 1
p=0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
r = bernoulli.rvs(p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()


```

```
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
```

```
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```
# binomial con cierto p entre 0 y 1 y cierto n
n,p = 10, 0.5
from scipy.stats import binom
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
r = binom.rvs(n, p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos binomialmente con p="+str(p)+" y n= "+str(n))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos binomial ordenados con p="+str(p)+" y n= "+str(n))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos binomialmente con p="+str(p)+" y n= "+str(n))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución binomial con p=",p,"y n=",n,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la distribución binomial con p=",p,"y n=",n,"son u_est=" ,avg,"var_est=",varest)
```

```
# Geométrica con cierto p entre 0 y 1
p = 0.3
from scipy.stats import geom
mean, var, skew, kurt = geom.stats(p, moments='mvsk')
r = geom.rvs(p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos geometricamente con p="+str(p))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos geometricamente con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos geométricamente con p="+str(p))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución geométrica con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para distribución geométrica con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```
# Poisson con cierto Lamnda L
L = 5
from scipy.stats import poisson
mean, var, skew, kurt = poisson.stats(L, moments='mvsk')
r = poisson.rvs(L, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos Poisson con L="+str(L))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos Poisson con L="+str(L))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos Poisson con L="+str(L))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución de Posisson con L=",L,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para distribución de Posisson con L=",L,"son u_est=" ,avg,"var_est=",varest)
```

```
#Binomial negativa con parametro p después de r fallos
r,p = 5,0.5
from scipy.stats import nbinom
mean, var, skew, kurt = nbinom.stats(r, p, moments='mvsk')
rn = nbinom.rvs(r, p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(rn)
plt.title(str(N)+" puntos distribuidos binomial negativa con p="+str(p)+" y r= "+str(r))
plt.show()
rn.sort()
plt.plot(rn)
plt.title(str(N)+" puntos distribuidos binomial negativa con p="+str(p)+" y r= "+str(r))
plt.show()
plt.hist(rn, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos binomial negativa con p="+str(p)+" y r= "+str(r))
plt.show()
print("Vamos a hallar algunas estimaciones de estadisticas de rn")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=rn.mean()
varest=rn.var()
print("Los valores teoricos de la media y varianza para la distribución binomial negativa con p=",p,"y r=",r,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la distribución binomial negativa con p=",p,"y r=",r,"son u_est=" ,avg,"var_est=",varest)
```

## Segundo punto Distribuciones contínuas.


```
#Para todas las simulaciones:
N=1000
m=300
numpy.random.seed(seed=1314926535)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```
# Normal de media 5 y desviación estandar 0.5
u=5
std=0.5
mean, var, skew, kurt = scipy.stats.norm.stats(moments='mvsk',loc=u,scale=std)
r = scipy.stats.norm.rvs (loc=u,scale=std, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales con media="+str(u)+" y std="+str(std))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normales con u="+str(u)+ " y std= "+str(std))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```
#T de Student con paramero K 
from scipy.stats import t
K=1.0
mean, var, skew, kurt = t.stats(moments='mvsk',df=K)
u=mean
std=var
r = t.rvs (df=K, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos t con media="+str(u)+" y var="+str(std))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos t ordenados con media="+str(u)+" y var="+str(std))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos t con u="+str(u)+ " y var= "+str(std))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```
# Exponencial con Lamnda L
from scipy.stats import expon
L=0.3
scl = 1 / L
mean, var, skew, kurt = expon.stats(moments='mvsk', scale=scl)
r = expon.rvs(size=N, scale=scl)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos exponencial con Lambda="+str(L))
plt.show()

r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos exponencial con Lambda="+str(L))
plt.show()


plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos exponencial con Lambda="+str(L))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución exponencial son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```

#Chi cuadrada con K grados de libertad
from scipy.stats import chi2
K=3.0
mean, var, skew, kurt = chi2.stats(K, moments='mvsk')
r = chi2.rvs(K, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos chi cuadrada con "+str(K)+" grados de libertad")
plt.show()

r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos chi cuadrada con "+str(K)+" grados de libertad")
plt.show()


plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos chi cuadrada con "+str(K)+" grados de libertad")
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución Chi cuadrada  son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```
#Gamma con parametros k y theta t
from scipy.stats import gamma
k,t=2.54,0.43
scl=1/t
mean, var, skew, kurt = gamma.stats(k, scale=scl, moments='mvsk')
r = gamma.rvs(k, scale=scl , size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos gamma con k="+str(k)+" y con t= "+str(t))
plt.show()

r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos gamma con k="+str(k)+" y con t= "+str(t))
plt.show()


plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos gamma con k="+str(k)+" y con t= "+str(t))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la distribución gamma  son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```
